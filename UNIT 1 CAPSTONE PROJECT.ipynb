## Experimental Design

#### Hypothesis: 
Individuals who quit smoking and live a healthier lifestyle will have lower BMI levels than those who do not quit smoking and live a healthier lifestyle.

#### Sample: 
We will choose 250 patients randomly from each of the four regions(Southeast, Southwest, Northeast, Northwest) that are smokers with BMI levels higher than 25. The total sample size would be 1000. We want patients that are willing to take part in this experiment in which half of the sample will live diet and excercise and quit smoking, and the other half will diet and excercise but continue to smoke. These patients will be asked to volunteer from hospitals in the US throughout the four regions.

#### Treatment:
We will have the treatment group be half of the amount of the total sample size. 500 people will be in the treatment group which will be the indivduals that will diet and exercise and quit smoking. The control group will also be 500 people who will diet and exercise, but they will not quit smoking.

## Rollout & Evaluation Plan

#### Impact Window: 
One month. In one month, it is possible for a person lose around 8-10 pounds in a month by living a healthier lifestyle so will use one month to determine if smoking has a significant effect on BMI.

#### Data Collection Approach: 
Once the patient decides they will participate in this experiment, they will have their first physical exam to record their BMI and plan a nutritiion plan and exercise routine. After the first physical evaluation, participants of this experiment will report to their local hospital for a physical examination after every two weeks. So in total, there will be three physical examinations to determine if there are any changes in BMI during the time of the experiment.

#### Metrics: 
The metrics used in this experiment will be the following: 
Age
BMI(body mass index) 
Weight(lbs), 
Number of ours of physical activity (per day) 
Calorie intake (per day) 
Number of cigarettes(per day for smokers)

#### Success Criteria: 
We are testing our hypothesis by determining if there is a significant difference in mean BMI levels between smokers and non smokers. By using a t-test, we will check obtain our p-value. If we get a p-value less than 0.05, then we can reject our null hpythesis, which means that there is a significant difference in BMI levels between smokers and nonsmokers who live a healthier lifestyle. This will support our claim that smoking doest affect a person's BMI.

If we get a p-value bigger than 0.05, then we can fail to reject our null hypothesis which will lead us to conclude that there would be no significant difference in BMI levels between smokers and nonsmokers who choose to live a healthier lifestyle.
